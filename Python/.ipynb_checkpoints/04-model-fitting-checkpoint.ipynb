# Developing a Random Forest Regressor to Predict Biomass (pgC per L)

### First need to features and variables that will be used for the RF model.

```python
covari.head(3)
```

```python
labels = covari.Biomass_pgC_per_L
labels
```

### Features will be used to predict the labels, or the biomass in our case.

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(labels)
# Remove the labels from the features
# axis 1 refers to the columns
droped_features = ['date','PopulationName','CellAbundance_10^6_cells_per_L',
                   'Biomass_pgC_per_L', 'CellAbundance_10^6_cells_per_L',
                  'CellDiameter_micrometer', 'cruisename', 'Light_micromolQuanta_m2_s',
                   'SizeClass','CellQuotas_fgC_per_cell']
features= covari.drop(droped_features, axis = 1)
# Saving feature names for later use
feature_list = list(features.columns)
# Convert to numpy array
features = np.array(features)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets"
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
# Looking at the shape of the features and labels to see if they match up
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# We are using a regressor RF model because we are predicting on continous values
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 100 decision trees
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
# Use the Training data to build the model
rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_labels, predictions)
print('Mean Absolute Error:', round(mae, 2), 'pgC per L.')
```

## Looking for feature importance to try and take out variables that are not useful 
### (and potentially add more data to train on as a result)

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

## The accuracy was negative, this could be due to extreme over-fitting due to an extremely small amount of data.

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
pd.DataFrame(rf.feature_importances_, index=feature_list).sort_values(by=0, ascending=False)

# Import matplotlib for plotting and use magic command for Jupyter Notebooks
import matplotlib.pyplot as plt
%matplotlib inline
# Set the style
plt.style.use('fivethirtyeight')
# list of x locations for plotting
x_values = list(range(len(importances)))
# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')
# Tick labels for x axis
plt.xticks(x_values, feature_list, rotation=45, ha='right', rotation_mode='anchor')

# Axis labels and title
plt.ylabel('Importance'); plt.title('Variable Importances');
#make a result with y as true biimass and x as predicted
#by tomorrow will have true vs predicted, by tuesday based on that plot, look at that graph and see if/whichh individual groups are doing well, then further imporviing my model
# can help me decide if I have one model for each population....
```

# Making a Random Forest Model to Predict Phytoplankton Size

### First need to features and variables that will be used for the RF model

```python
covari.head() # predict cell abundance, biomass, or cell quotas
# biomass is a product of cell abundance and cell quotas
# go biomass
```

```python
labels = covari.PopulationName
# One-hot encode the data using pandas get_dummies
labels = pd.get_dummies(labels)
# Display the first 5 rows of the last 12 columns
labels.head(5)
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(labels)
# Remove the labels from the features
# axis 1 refers to the columns
droped_features = ['date','PopulationName','CellAbundance_10^6_cells_per_L',
                   'Biomass_pgC_per_L', 'CellAbundance_10^6_cells_per_L',
                  'CellDiameter_micrometer', 'cruisename', 'Light_micromolQuanta_m2_s',
                  'SizeClass','CellQuotas_fgC_per_cell']
features= covari.drop(droped_features, axis = 1)
# Saving feature names for later use
feature_list = list(features.columns)
# Convert to numpy array
features = np.array(features)
```

```python
feature_list
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets"
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
# Looking at the shape of the data to see if this was done right
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```
